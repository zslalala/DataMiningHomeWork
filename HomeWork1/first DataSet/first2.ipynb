```python
import os
import pandas as pd
import numpy as  np
import matplotlib.pyplot as plt

def load_data():
    csv_path = "cbg_patterns.csv"
    return pd.read_csv(csv_path)
data=load_data()
data.hist(bins=15, figsize=(15, 20))
plt.show()
```

接下来对各个数值属性进行QQ图检验其是否服从正态分布，选取census_block_group属性进行检测

```python
from scipy import stats
stats.probplot(data["census_block_group"], dist="norm", plot=plt)
plt.show()
```

```python
import seaborn as sns
sns.set(style="whitegrid")
sns.boxplot(data=data["census_block_group"])
plt.show()
```
