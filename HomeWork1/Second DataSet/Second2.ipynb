```python
import pandas as pd
import matplotlib.pyplot as plt
def load_data():
    csv_path = "winemag-data-130k-v2.csv"
    return pd.read_csv(csv_path)
data = load_data()
data.hist(bins=15, figsize=(15, 20))
plt.show()
```

选取point属性绘制QQ图观察正态分布情况

```python
from scipy import stats
stats.probplot(data["points"], dist="norm", plot=plt)
plt.show()
```

结果拟合为直线，说明points属性具有较好的正态分布情况，然后对其进行盒图测试，观察离散情况

```python
import seaborn as sns
sns.set(style="whitegrid")
sns.boxplot(data=data["points"])
plt.show()
```
